```python
import pickle

training_file  = "./data/train.p"
testing_file = "./data/test.p"

with open(training_file,mode = 'rb') as f:
    train = pickle.load(f)
with open(testing_file,mode = 'rb') as f:
    test = pickle.load(f)
    
x_train,y_train = train['features'],train['labels']
x_test,y_test = test['features'],test['labels']

print("X_train shape:", x_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", x_test.shape)
print("y_test shape:", y_test.shape)
```

```python
class_names = ['20','30','50','60','70','80','no 80','100',
              '120','no passing','car and truck','baby','yellow diamond',
              'yield','stop','white circle','truck','one way','dnager ahead','left curve','right curve',
              'straight-left-straight','bumps','skidding road','narrow lane',
              'contruction/work','traffic signal','pedestrian','people crossing',
              'no cycling','ice','animals cross','national speed limit applies',
              'right turn','left turn','straight','straight and right','striaght and left',
              'keep right','keep left','roundabout','no overtaking','no car or truck']
```

```python
x_train = x_train/255.0
x_test=x_test/255.0
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import random

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    r = random.randint(0,39000)
    plt.imshow(x_train[r],cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[r]])
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(128,activation=tf.nn.relu),
    keras.layers.Dense(43,activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer=tf.train.AdamOptimizer(),
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=6)
```

```python
test_loss,test_acc = model.evaluate(x_test,y_test)
print('Test Accuracy:',test_acc)
```
